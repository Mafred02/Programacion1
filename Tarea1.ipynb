```python
l = [1, 2, 3, 4, 5, 6]
t = tuple(l)
d = dict(enumerate(l))
c = set(l)
print(l, t, d, c)
```

```python
t = (7, 8, 9, 10, 11, 12)
d = dict(enumerate(t))
c = set(t)
l = list(t)
print(t, d, c, l)
```

```python
d = {
    "nombre": "Nicolas",
    "apellido": "Rodriguez",
    "edad": 17
}
c = set(d.items())
l = list(d.items())
t = tuple(d.items())
print(d, c, l, t)
```
